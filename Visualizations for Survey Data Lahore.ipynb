```python
# create a list of our conditions
conditions = [
    (violence_dframe['res_prf_age'] >= 18) &
    (violence_dframe['res_prf_age'] < 30), 
    (violence_dframe['res_prf_age'] >= 30) & 
    (violence_dframe['res_prf_age'] < 50),
    (violence_dframe['res_prf_age'] >= 50) &
    (violence_dframe['res_prf_age'] < 70),
    (violence_dframe['res_prf_age'] >= 70) &
    (violence_dframe['res_prf_age'] <= 100)
    ]

# create a list of the values we want to assign for each condition
values = ['Under 30', 'Between 30 and 50','Between 50 and 70','Between 70 and 100']

# create a new column and use np.select to assign values to it using our lists as arguments
violence_dframe['agecategory'] = np.select(conditions, values)
```
